The notion of a system is central in digital communications and particularly system's theory. Abstractly, a system is defined as something that takes an input signal and produces an output signal by some transformation rule Tr

<p style="text-align: center;">y(t)=Tr{x(t)}</p>



```python
import numpy as np
import matplotlib.pyplot as plt
```

A system can amplify the input signal, by doubling its amplitude:

<p style="text-align: center;">y(t)=2x(t)</p>

We take the input signal.

<p style="text-align: center;">x(t)=sin(t)</p>

Then, we can straight-forwardly implement the system as taking a function (a signal) and returning another time function (another signal):

```python
x = lambda t: np.sin(t)

def Tr(x):
    return lambda t: 2*x(t)
```

Plot the input signal and the output signal over some time:

```python
t = np.linspace(-5,5,1000)
plt.plot(t, x(t), label='Input $x(t)$')
y = Tr(x)
plt.plot(t, y(t), label='Output $y(t)=2x(t)$')
```
